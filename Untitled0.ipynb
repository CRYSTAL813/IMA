```python
import numpy as np
import pandas as pd
fashion_df = pd.read_csv(r"D:/Anaconda/python/fashion.csv")
fashion_df
```

```python
fashion_df["Gender"].value_counts()
```

```python
shoe_df=fashion_df[fashion_df["Category"].str.contains("Apparel")==False]
shoe_df.reset_index()
```

```python
Y = shoe_df['Gender']
```

```python
import tensorflow.keras as K 
import tensorflow as tf
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(shoe_df,Y,test_size=0.2)
x_train, x_val, y_train, y_val = train_test_split(X_train,Y_train,test_size=0.125)
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rotation_range = 30,
                                   zoom_range = 0.2,
                                   width_shift_range = 0.1,
                                   height_shift_range = 0.1,
                                   horizontal_flip = True,
                                   rescale = 1./255) 
val_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
#batch_size = 100
#training_set = train_datagen.flow_from_directory(train_dir,target_size = (64, 64), batch_size = batch_size) 
#val_set = val_datagen.flow_from_directory(validate, target_size = (64, 64), batch_size = batch_size)
#test_set = test_datagen.flow_from_directory(test_dir,
                                            #target_size = (64, 64),
                                            #batch_size = batch_size)
training_set=train_datagen.flow_from_dataframe(dataframe=x_train, directory="Footwear/Footwear/both/Images/images_with_product_ids", x_col="Image", y_col="Gender", class_mode = 'binary', target_size=(64,64), batch_size=50)
val_set=val_datagen.flow_from_dataframe(dataframe=x_val, directory="Footwear/Footwear/both/Images/images_with_product_ids", x_col="Image", y_col="Gender", class_mode = 'binary', target_size=(64,64), batch_size=50)
test_set=test_datagen.flow_from_dataframe(dataframe=X_test, directory="Footwear/Footwear/both/Images/images_with_product_ids", x_col="Image", y_col="Gender", class_mode = 'binary', target_size=(64,64),batch_size=50)

    
```

```python
from keras.models import Sequential 
from keras.layers import Dense, Dropout, Flatten, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D 

cnn = Sequential()

cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(64, 64, 3)))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='SAME'))
cnn.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='SAME'))
cnn.add(Flatten())
cnn.add(Dense(64, activation='relu'))
cnn.add(Dropout(0.25))
cnn.add(Dense(1, activation='sigmoid'))
```

```python
cnn.summary()
```

```python
from keras import losses

cnn.compile(optimizer = 'adam', 
            loss = losses.BinaryCrossentropy(), 
            metrics = ['accuracy'])

history = cnn.fit(training_set, validation_data = val_set, epochs = 10)
```
